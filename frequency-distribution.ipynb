```python
text="I saw John coming. He was with Mary. I talked to John and Mary. \
John said he met Mary on the way. John and Mary were going to school."

#using nltk
tokens = nltk.word_tokenize(text)

print(tokens)
print()

#using spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

token = [str(token) for token in doc]
print(token)

print()

# Create the frequency distribution
freq_dist = nltk.FreqDist(token)

print(freq_dist.most_common())

print()
# Print the most common words and their frequencies
for word, frequency in freq_dist.most_common():
    print(word, frequency)
```
