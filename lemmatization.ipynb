```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
word_list = ["friend", "friendship", "friends", "friendships"]
```

```python
'''
POS- 
Noun-n
verb-v
adjective-a
adverb-r
'''

for word in word_list:
    print('{0:20} {1:20}'.format(word,lemma.lemmatize(word,pos = 'n')))
```

```python
import spacy

# Load the language model
nlp = spacy.load("en_core_web_sm")
```

### `If you require a larger and more accurate model, you can use en_core_web_lg instead.`

```python
word_list = ["friend", "friendship", "friends", "friendships"]
```

```python
lemmas = []
for word in word_list:
    doc = nlp(word)
    print(doc)
    lemmas.append(doc[0].lemma_)
```

```python
for word,lemma in zip(word_list,lemmas):
    print('{0:20}-> {1:20}'.format(word,lemma))
```

```python
text = "I am running in the park and saw many beautiful birds."
doc = nlp(text)
print(doc)

token = [tokens.lemma_ for tokens in doc]
print(token)
```
