```python
df = pd.DataFrame({'reviews' : [
    "This movie is amazing!",
    "The plot was confusing and the acting was terrible.",
    "I enjoyed the movie. The characters were well-developed.",
    "I didn't like the movie at all. It was boring and predictable."
]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(ngram_range=(1,1)) #bog of words(default)
cv1 = CountVectorizer(ngram_range=(1,2)) #uni-grams and bi-grams
cv2 = CountVectorizer(ngram_range=(2,2)) #bi-grams
cv3 = CountVectorizer(ngram_range=(3,3)) #tri-grams
```

```python
x = df['reviews']

x_cv = cv.fit_transform(x)
x_cv1 = cv1.fit_transform(x)
x_cv2 = cv2.fit_transform(x)
x_cv3 = cv3.fit_transform(x)
```
