# Using NLTK library

```python
import nltk
from nltk.tokenize import (word_tokenize,
                          sent_tokenize,
                          TreebankWordTokenizer,
                          wordpunct_tokenize,
                          TweetTokenizer,
                          MWETokenizer)
```

```python
text = 'Hope, is the only thing stronger than fear! #Hope'
print(word_tokenize(text))
```

```python
text1 = 'I have a Ph.D in A.I'
text2 = "We're here to help mail us at abc@gmail.com"
text3 = 'A 5km ride cost $10.75'
```

```python
word_tokenize(text1)
```

```python
word_tokenize(text2)
```

```python
word_tokenize(text3)
```

## Punctuation tokenization

```python
text = 'Hope, is the only thing stronger than fear! #Hope'
print(wordpunct_tokenize(text))
```

## TreeBank Word Tokenizer

```python
text = "What you don't want to be done to yourself, do'nt do to others 100.0%üëç..."
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(text))
```

# Using Spacy
    -Here we have to convert text to document

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc1 = nlp(text1)
doc2 = nlp(text2)
doc3 = nlp(text3)
```

```python
for token in doc1:
    print(token)
```

```python
for token in doc2:
    print(token)
```

```python
for token in doc3:
    print(token)
```

```python
word = 'Dr. Strange makes villan to zero with his magic. Hulk makes villan to zero with his power'
doc = nlp(word)

for token in doc.sents:
    print(token)
    
print()
print()

for sentence in doc.sents:
    for word in sentence:
        print(word)
```
