<h1 style = 'text-align:center'>Stemming</h1>

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem_words(text):
    
    return ' '.join([ps.stem(word) for word in text.split()])
```

```python
stem_words('walks walker walked walking')
```

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
for word in words:
    print(word,'----->',ls.stem(word)) 
```

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english',ignore_stopwords = True)
```

```python
words=["eating","eats","eaten","writing","writes","programming",'the',"programs","history","finally","finalized",'is']
```

```python
for word in words:
    print(word,'---->',ss.stem(word))
```

# PorterStemmer vs LancasterStemmer vs SnowballStemmer

```python
word_list = ["friend", "friendship", "friends", "friendships","eating","eats","eaten","writing","writes","programming",'the',"programs","history","finally","finalized",'is']

ps = PorterStemmer()
ls = LancasterStemmer()
ss = SnowballStemmer('english')
```

```python
print('{0:20} {1:20} {2:20} {3:20}'.format('Orgianl Word','PorterStemmer','LancasterStemmer','SnowballStemmer'))
for word in word_list:
    print('{0:20} {1:20} {2:20} {3:20}'.format(word,ps.stem(word),ls.stem(word),ss.stem(word)))
```
